```python
import pandas as pd
churn_df = pd.read_csv("churn_scores.csv")
churn_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

churn_df["tier"] = pd.cut(churn_df["lifetime_value"], 
                          bins=[0, 200, 500, 1000, float("inf")], 
                          labels=["bronze", "silver", "gold", "platinum"])

plt.figure(figsize=(8, 5))
sns.boxplot(data=churn_df, x="tier", y="churn_risk_score", palette="coolwarm")
plt.title("Churn Risk by Customer Tier")
plt.xlabel("Tier")
plt.ylabel("Churn Risk Score")
plt.show()
```

```python
anomalies_df = pd.read_csv("weekly_anomalies.csv")

plt.figure(figsize=(12, 6))
plt.plot(anomalies_df["week_start"], anomalies_df["total_amount"], label="Revenue")
plt.scatter(
    anomalies_df[anomalies_df["is_anomaly"] == 1]["week_start"],
    anomalies_df[anomalies_df["is_anomaly"] == 1]["total_amount"],
    color="red", label="Anomaly"
)
plt.title("Weekly Revenue with Anomaly Detection")
plt.xlabel("Week Start")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
